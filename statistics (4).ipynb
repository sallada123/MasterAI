# Discrete Random Variable

A random variable is said to be discrete if it can take a countable number of discrete values. This means that
it is often represented in the form of whole numbers ( 1,2,3...).  

Examples of discrete random variables are the number of children in a family, the number of coin tosses , and the 
number customers in a grocery store ,etc.

We can observe how we cannot use fractions or decimals to represent the above quantities. In this manner , we can 
differentiate between a discrete and continuous random variable.

<h1 style="font-size:3em;color:green;">Let us talk about Poisson Distribution</h1>

 A **Poisson distribution** is a discrete distribution that describes the number of events occuring in a fixed interval of time or region of opportunity.

Let a Poisson Random Variable **X** be the number of successes that occur in a time period that occurs with average rate *λ* (in that time period):

Keywords that will *mostly* help us spot Poisson distributions in problems = per day,week,month or year... (**time period**)

<u>Probability mass function</u>: 
            $P(X=x)=\frac{e^{-\lambda} \lambda^{x}}{x!}$

where 
* e= Euler's number(Napier's constant ≈ 2.72 (We only need to remember the value)
* x= 1,2,3..
* λ= The average rate 

<u>Assumptions that might help us differentiate Poisson distributions from the other distributions </u>:
* The rate at which the events occur must be constant:
  The probability of an event occuring in a certain time interval must be exactly the same for every other time interval of   the same length.
  
* The occurrence of one event doesn't affect the occurence of a subsequent event.

# Let us now look at histograms for the different values of λ

## When the value of λ(the mean)=1:

```python
from scipy.stats import poisson
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_rvs=pd.Series(poisson.rvs(1,size=10000,random_state=2))
```

```python
data=x_rvs.value_counts().sort_index().to_dict()
data
```

```python
fig,ax=plt.subplots(figsize=(19,5))
ax.bar(range(len(data)),list(data.values()),align='center')
plt.xlabel("PoissonRV")
plt.ylabel("Frequency")
plt.show()
```

## Now let us solve a few problems to get a better understanding of the Poisson Distribution

**Example 1**: Let $Y$ denote a random variable that has a Poisson distribution with mean($\lambda$)=2.

To find:

a.P(Y=3)

b.P(Y<3)

c.P(Y$>=$3)





**Solutions with explanations**:

 $P(Y=y)=\frac{e^{-\lambda} \lambda^{y}}{y!}$
 
 a.
 $P(Y=3)=\frac{e^{-2} \lambda^{3}}{3!}$
        =$\frac{2.72^{-2} 2^{3}}{3!}$
        =$0.18$
        
 b. We know that the lower limit for the Poisson Distribution is 0.
 Therefore,
 
 $P(Y<3)=P(Y=0)+P(Y=1)+P(Y=2)$
        =$\frac{2.72^{-2} 2^{0}}{0!}$+$\frac{2.72^{-2} 2^{1}}{1!}$+$\frac{2.72^{-2} 2^{2}}{2!}$
        =$0.135+0.2703+0.2703$
        =$0.6756$
        =$0.676$
        

c. $P(Y>=3)$

Since,
$P(Y>=3)+P(Y<3)=1$
$P(Y>=3)=1-P(Y<3)$
        = $1-0.676$
        = $0.324$

**Example 2**:
    A salesperson has found that the probability of a sale on a single contact is approximately 0.03. If the salesperson contracts 100 prospects, what is the approximate probability of making at least one sale? 

Note:Eventhough, the question doesn't mention any particular time period,it's still a Poisson distribution.This is because the time period would be the amount of time taken to contract 100 prospects.

**Solution:**

Let $Y$ be the number of sales on $100$ contracts.

Using the Poisson approximation,
                                $\lambda = np$ = $ 0.03 * 100$ = $3$

To find: 
$P(Y>=1)$

$P(Y>=1)= 1-P(Y=0)$
        = $ 1- \frac{e^{-3} 3^{0}}{0!}$
        ≈$0.95$
